## The power of COLUMN TRANSFORMERS

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Credit Score']),df['Credit Score'],
                                                test_size=0.2)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1', OrdinalEncoder(categories=[['Bachelor', 'High School','Master', 'PhD'],['Yes', 'No']]), ['Education Level','Has Vehicle']),
    ('tnf2',OneHotEncoder(sparse_output=False,drop='first'),['Gender','City','Job Title','Marital Status',])
],remainder='passthrough')
```

```python
transformed_data = transformer.fit_transform(df)
```

```python
X_train_transformed=transformer.fit_transform(X_train)
```

```python
transformer.transform(X_test).shape
```
