```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df_e = df.iloc[:,5:6]
```

```python
df_e
```

```python
oe = OrdinalEncoder(categories=[['Bachelor', 'High School','Master', 'PhD']])
```

```python
oe.fit(df_e)
```

```python
X_train_e = oe.transform(df_e)
```

```python
X_train_e
```

```python
oe.categories_
```

## One hot encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)
```

```python
df.sample(3)
```

```python
df_oh = df[['Gender','City','Job Title','Marital Status']]
```

```python
ohe_new = ohe.fit_transform(df_oh)
```

```python
ohe_new 
```
