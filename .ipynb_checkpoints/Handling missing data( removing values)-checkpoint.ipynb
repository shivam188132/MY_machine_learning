### Can't perform CCA on brand colulmn 

```python
cols = [var for var in df.columns if df[var].isnull().mean()*100 < 5 and df[var].isnull().mean() > 0]
cols
```

```python
df[cols].sample(3)
```

```python
len(df[cols].dropna())/len(df[cols])*100
```

```python
new_df = df[cols].dropna()
df[cols].shape, new_df.shape
```

```python
plt.figure(figsize=(8,4))
plt.subplot(121)
sns.distplot(new_df["retail_price"])
plt.title('New data frame PDF')
```

```python
new_df.describe()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(new_df["retail_price"], kde=True, log_scale=(True, False))
plt.title("New data frame PDF with Log Scale on X-axis")
plt.xlabel("Retail Price (Log Scale)")
plt.ylabel("Relative frequencies")
plt.show()
```

```python
# Subplot 2: Distribution filtered by a range
plt.subplot(1, 2, 2)
sns.histplot(new_df[new_df["retail_price"] < 5000]["retail_price"], kde=True, bins=50)
plt.title("Retail Price Distribution (Filtered < 10,000)")

plt.tight_layout()
plt.show()
```
