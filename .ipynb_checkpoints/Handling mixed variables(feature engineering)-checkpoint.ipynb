```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
df['number'].value_counts().plot(kind='bar')
plt.title('Passengers travelling with', fontsize=14, color='blue', loc='center')
```

```python
df.sample(4)
```

```python
df['number( numeric)'] = df['number'].str.extract('(\d+)')# gives result as a object dtype
df['number(alpha)'] = df['number'].str.extract('([A-Za-z]+)')
df
```

```python
df.dtypes
```

```python
#to extract numbers in integer format
df['number_numerical'] = pd.to_numeric(df["number"],errors='coerce',downcast='integer')
df.dtypes
```

```python
df['Cabin (NUM)'] = df['Cabin'].str.extract('(\d+)') #captures numerical part
df['Cabin (ALPHA)'] = df['Cabin'].str[0] #captures first letter
df
```

```python
# Cabin (NUM)-->  object
# Ticket(NUM)  -->  object
```
